Libraries Imported

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
```

Loading Data

```python
data = pd.read_excel('ANTM.xlsx', sheet_name='data', header=None, skiprows=1, usecols="A:F")
X = data.iloc[:, :5].values.T
T = data.iloc[:, 5].values.T
t = np.arange(1, len(T) + 1)
```

```python
print(t)
```

Handling Missing Values

```python
T[np.isnan(T)] = np.nanmedian(T)
```

Neural Network Training Loop

```python
results = pd.DataFrame(columns=['h', 'RMSE', 'T', 'Yt', 'Decision'])

# Neural Network Training Loop
for h in range(1, 200):
    # Create and train neural network
    net = MLPRegressor(hidden_layer_sizes=(h,), solver='lbfgs', max_iter=1000)
    net.fit(X.T, T)

    # Predict on training data
    Y = net.predict(X.T)

    # Calculate RMSE
    RMSE = np.sqrt(mean_squared_error(T, Y))

    # Predict on test data
    Xt = pd.read_excel('ANTM.xlsx', sheet_name='data', header=None, skiprows=150, usecols="A:E").values.T
    Yt = net.predict(Xt.T)[0]

    # Make a decision
    if Yt > T[-1]:
        Decision = 'buy'
    elif Yt == T[-1]:
        Decision = 'Hold'
    else:
        Decision = 'Sell'

    # Selects the last T value
    single_T_value = T[-1]

    # Store results in the DataFrame
    results = results.append({'h': h, 'RMSE': RMSE, 'T': single_T_value, 'Yt': Yt, 'Decision': Decision}, ignore_index=True)

    # Print results for each iteration
    tab = pd.DataFrame({'t': t, 'T': T, 'Y': Y, 'e': T - Y})
```

Saving Results

```python
results.to_excel('Results.xlsx', index=False)
```
